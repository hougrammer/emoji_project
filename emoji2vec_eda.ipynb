### Barbieri emoji vectors: The dimensions of the vectors is 300. raw is the raw model, where we trained the embedding on the whole dataset (only links are removed), the clean embeddings were trained on a cleaner dataset: stopwords, punctuation and links were removed. onlyemo are embedding trained only on emojis. In these models, the emojis character have been replaced by eojiCODE where CODE is the unicode of the emoji.

```python
barbieri_raw = gensim.models.KeyedVectors.load_word2vec_format('data/raw_model_swm-300-6-3.w2v')
barbieri_clean = gensim.models.KeyedVectors.load_word2vec_format('data/clean_model_swm-300-6-3.w2v')
barbieri_onlyemo = gensim.models.KeyedVectors.load_word2vec_format('data/onlyemo_model_swm-300-3-3.w2v')
```

```python
print(len(e2v.vocab))
print(len(barbieri_raw.vocab))
print(len(barbieri_clean.vocab))
print(len(barbieri_onlyemo.vocab))
```

```python
print(u'\U0001f604')
print(barbieri_raw['eoji1f604'])
print(barbieri_clean['eoji1f604'])
print(barbieri_onlyemo['eoji1f604'])
```

```python
print(sigmoid(np.matrix(barbieri_raw['eoji1f604'])*np.transpose(np.matrix(w2v['smile']))))
print(sigmoid(np.matrix(barbieri_clean['eoji1f604'])*np.transpose(np.matrix(w2v['smile']))))
print(sigmoid(np.matrix(barbieri_onlyemo['eoji1f604'])*np.transpose(np.matrix(w2v['smile']))))
```

```python
top_5_raw = pd.DataFrame(np.matrix(w2v.vectors)*np.transpose(np.matrix(barbieri_raw['eoji1f604'])),index = w2v.vocab).apply(sigmoid, axis = 1)
top_5_clean = pd.DataFrame(np.matrix(w2v.vectors)*np.transpose(np.matrix(barbieri_clean['eoji1f604'])),index = w2v.vocab).apply(sigmoid, axis = 1)
top_5_onlyemo = pd.DataFrame(np.matrix(w2v.vectors)*np.transpose(np.matrix(barbieri_onlyemo['eoji1f604'])),index = w2v.vocab).apply(sigmoid, axis = 1)
```
