```python
tf.reset_default_graph()

chkp.print_tensors_in_checkpoint_file('./simple-examples/models/model.ckpt-2721',tensor_name='', all_tensors=True)

with tf.Session() as sess:
    new_saver = tf.train.import_meta_graph('./simple-examples/models/model.ckpt-2721.meta')
    new_saver.restore(sess, tf.train.latest_checkpoint('./simple-examples/models/'))
```
